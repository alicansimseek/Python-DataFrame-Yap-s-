```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("applications(2.0).csv")
```

```python
dataset
```

```python
dataset.applicant_id.duplicated().sum()                           #applicant_id sütunundaki tekrar eden değerleri kaldırın.
```

```python
dataset.drop_duplicates(subset = ["applicant_id"],keep="first",inplace=True)
```

```python
dataset
```

```python
dataset["External Rating"].fillna(0,inplace=True)    #'External Rating' alanındaki eksik değerleri sıfırla doldurun.
```

```python
dataset["Education level"].fillna("Ortaokul",inplace=True)
```

```python
dataset2 = pd.read_csv("industries(2.0).csv")
```

```python
dataset2
```

```python
df = dataset.merge(dataset2,how="left") #2. Bu DataFrame'e industries.csv dosyasından gelen verileri, yani sektörlerin derecelendirmelerini ekleyin.
```

```python
df
```

```python
((df['Age'] >= 35) & (df['Age'] <= 55)) * 20
```

```python
df['Score2'] = (
    # Yaş kriteri
    ((df['Age'] >= 35) & (df['Age'] <= 55)) * 20 +
    # Hafta içi başvuru kriteri
    (~df['Applied at'].isin(['Saturday', 'Sunday'])) * 20 +
    # Medeni durum kriteri
    (df['Marital status'] == 'Married') * 20 +
    # Lokasyon kriteri
    (df['Location'] == 'Kiev') * 10 +
    # Sektör skoru
    df['Score'] +
    # External Rating kriterleri
    (df['External Rating'] >= 7) * 20 -
    (df['External Rating'] <= 2) * 20
)

# Eksik Amount veya sıfır External Rating durumunda puanı sıfıra ayarlama
df.loc[df['Amount'].isna() | (df['External Rating'] == 0), 'Score2'] = 0

# Puanı 0-100 arasında sınırlama
df['Score2'] = df['Score2'].clip(0, 100)
df.loc[df['Amount'].isna() | (df['External Rating'] == 0), 'Score2'] = 0


```

```python
df
```

```python
df.drop(df[df["Score2"] <= 0].index)
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
# "Applied at" sütununu datetime formatına çeviriyoruz (eğer zaten değilse)
df['Applied at'] = pd.to_datetime(df['Applied at'])

# Başvuruları haftaya göre gruplayıp, Score2'nin ortalamasını alıyoruz
df['Week'] = df['Applied at'].dt.isocalendar().week  # Yıl ve hafta bilgisi alınıyor
grouped_data = df[df["Score2"] > 0].groupby('Week')['Score2'].mean().reset_index()
```

```python
# Haftalık ortalama Score2'yi çiziyoruz
plt.figure(figsize=(10,6))
plt.plot(grouped_data['Week'], grouped_data['Score2'], marker='o', linestyle='-', color='b')

# Başlık ve etiketler
plt.title('Haftalık Ortalama Kabul Edilen Başvuru Puanları')
plt.xlabel('Hafta')
plt.ylabel('Ortalama Puan')

# Grafik gösterimi
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

plt.hist(grouped_data['Score2'], bins=10, color='blue')

plt.title('Haftalık Ortalama Kabul Edilen Başvuru Puanlarının Dağılımı')
plt.xlabel('Ortalama Puan')
plt.ylabel('Frekans')

# Grafik gösterimi
plt.grid(True)
plt.tight_layout()
plt.show()
```
