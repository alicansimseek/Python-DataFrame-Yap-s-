```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("applications(2.0).csv")
```

```python
dataset
```

```python
dataset.applicant_id.duplicated().sum()                           #applicant_id sütunundaki tekrar eden değerleri kaldırın.
```

```python
dataset.drop_duplicates(subset = ["applicant_id"],keep="first",inplace=True)
```

```python
dataset
```

```python
dataset["External Rating"].fillna(0,inplace=True)    #'External Rating' alanındaki eksik değerleri sıfırla doldurun.
```

```python
dataset["Education level"].fillna("Ortaokul",inplace=True)
```

```python
dataset2 = pd.read_csv("industries(2.0).csv")
```

```python
dataset2
```

```python
df = dataset.merge(dataset2,how="left") #2. Bu DataFrame'e industries.csv dosyasından gelen verileri, yani sektörlerin derecelendirmelerini ekleyin.
```

```python
df
```

```python
((df['Age'] >= 35) & (df['Age'] <= 55)) * 20
```

```python
df['Score2'] = (
    # Yaş kriteri
    ((df['Age'] >= 35) & (df['Age'] <= 55)) * 20 +
    # Hafta içi başvuru kriteri
    (~df['Applied at'].isin(['Saturday', 'Sunday'])) * 20 +
    # Medeni durum kriteri
    (df['Marital status'] == 'Married') * 20 +
    # Lokasyon kriteri
    (df['Location'] == 'Kiev') * 10 +
    # Sektör skoru
    df['Score'] +
    # External Rating kriterleri
    (df['External Rating'] >= 7) * 20 -
    (df['External Rating'] <= 2) * 20
)

# Eksik Amount veya sıfır External Rating durumunda puanı sıfıra ayarlama
df.loc[df['Amount'].isna() | (df['External Rating'] == 0), 'Score2'] = 0

# Puanı 0-100 arasında sınırlama
df['Score2'] = df['Score2'].clip(0, 100)
df.loc[df['Amount'].isna() | (df['External Rating'] == 0), 'Score2'] = 0


```

```python
df
```

```python
df.drop(df[df["Score2"] <= 0].index)
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
# Haftalık gruplama için 'Applied at' sütununu datetime formatına çeviriyoruz
df['Applied at'] = pd.to_datetime(df['Applied at'])
df['Week'] = df['Applied at'].dt.strftime('%Y-%U')  # Haftalık format: Yıl-Hafta

# Her haftanın kabul edilen başvurularının ortalama puanını hesaplıyoruz
weekly_avg_score = df.groupby('Week')['Score2'].mean().reset_index()

# Ortalamaları azalan sırada sıralıyoruz
weekly_avg_score = weekly_avg_score.sort_values(by='Week')

# Görselleştirme
plt.figure(figsize=(10, 6))
bars = plt.bar(weekly_avg_score['Week'], weekly_avg_score['Score2'], color=plt.cm.coolwarm(range(len(weekly_avg_score))))

# Her barın üstüne değer eklemek
for bar, score in zip(bars, weekly_avg_score['Score2']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), f'{score:.1f}', ha='center', va='bottom')

# Grafik ayarları
plt.title('Haftalara Göre Ortalama Başvuru Puanları', fontsize=14)
plt.xlabel('Hafta', fontsize=12)
plt.ylabel('Ortalama Puan', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
